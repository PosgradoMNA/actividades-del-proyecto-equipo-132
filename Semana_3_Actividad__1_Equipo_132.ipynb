# DATA WAREHOUSING
El almacenamiento de datos (“Data warehousing”) es el proceso mediante el cual se construye una base de datos gigante con información proveniente de muchas fuentes. Esto genera un desafío pues para lograr que una consulta sea provechosa y nos permita obtener la información necesaria para realizar determinados análisis, los datos tienen que ser normalizados independientemente de la fuente de donde provengan. El almacenamiento de datos usualmente se lleva a cabo creando un modelo que clasifica las dimensiones de los datos, una dimensión es una propiedad o característica de la entidad en la que está basada los datos. 



Existen dos conceptos para efectuar la normalización de datos, la más antigua con sus siglas en inglés es ETL, traduciendo significan extraer, transformar y cargar. En la etapa de transformación es donde se lleva a cabo el proceso de normalización, pero siguiendo este concepto al ser este un proceso intermedio para cada uno de los datos de entrada, se necesitan de una gran cantidad de recursos de procesamiento para trabajar con Big Data. Para resolver esta problemática es que aparece el concepto de ELT (extraer, cargar, transformar) que se caracteriza porque el usuario realiza la descarga de información en “bruto” y el proceso de normalización o transformación se realiza de manera virtual con herramientas y computadoras que se encuentran en la nube.

```python
# Verificar si hay datos nulos o datos faltantes en el dataset
df.isnull().values.any()
```

```python
# Verificar que columnas dentro del dataframe tienen datos nulos
df.isnull().any()
```

```python
# Definir la columna ID como el index de cada registro (fila)
df=df.set_index('ID')
```

```python
# Se obtiene la suma de datos nulos en cada columna ordenados de menor a mayor
df.isnull().sum().sort_values()
```

```python
# It1:Se inicia la limpieza del dataset elimiminando una cantidad no significativa de registros 
# Se observa que X2,X3,X4 y Y contienen variables categóricas y que hay muy pocos registros con datos nulos comparado con el 
# total de registros que es de 30,000
dfIt1= df.dropna(axis=0,subset=["X2","X3","X4","Y",])
```

```python
#Después de correr limpieza It1, se observa que en la columna "Y" ya no existen registros nulos pero aún hay otras columnas Xn
#con datos de entrada con datos nulos
dfIt1.isnull().sum().sort_values()
```

```python
#Suma total de filas con al menos 1 dato nulo después de limieza It1
suma1_col_It1=dfIt1.isnull().sum()
suma1_col_It1.sum()
```

```python
dfIt1.info()
```

```python
#Con la intención de no afectar la respuesta del modelo "Y" con suposiciones de imputación y tomando en cuenta que ya solo quedan 124 filas de
# un total de 29,995 es decir un 0.413% se decide eliminar todas las filas restantes con datos nulos

(100* (suma1_col_It1.sum()))/len(dfIt1)
```

```python
#Eliminando las 124 filas restantes con datos nulos
dfIT2=dfIt1.dropna(axis=0)
```

```python
#Como resultado final se tienen 29,958 registros reportando información en cada una de las variables Xn y en la respuesta de 
#salida Yn
dfIT2.isnull().any()
```

```python
#Verifiar total de registros después de eliminar datos nulos
dfIT2.info()
```

```python
#Se observa que X3 y X4 tiene mínimos o máximos fuera del rango descrito en la información del dataset. X3 muestra registros con < 0 y > 4 y X4 muestra registros < 0.
#Se hará un conteo de registros afectados para tomar decisión
dfIT2.describe()
```

```python
#X3 se observa que hay 311 registros que exceden el límite superior de 4 y hay 14 registros por debajo de límite inferior 1
#No es conveniente eliminar estos registros debido a la catidad, son 345.
dfIT2['X3'].value_counts() 
```

```python
#Se define una función para cambiar los valores fuera de rango de la dimension X3, si el valor es >4 se asigna el valor 4 y si el valor es < 1 se asigna el valor 1
def min_max_ed (value):
  if value>4:
    return 4
  elif value<1:
    return 1
  else:
    return value
```

```python
#Se reliza el cambio de valores de la dimension X3 y actualización dentro del dataframe dfIT2
dfIT2['X3']=dfIT2['X3'].apply(min_max_ed)
dfIT2['X3'].value_counts() 
```

```python
#X4 se observa que hay 54 registros por debajo del límite inferior que es 1
#No es conveniente eliminar estos registros debido cantidad, son 54
dfIT2['X4'].value_counts() 
```

```python
#Se define una función para cambiar los valores fuera de rango de la dimension X4, si el valor es < 1 se asigna el valor 1
def min_marr (value):
  if value<1:
    return 1
  else:
    return value
  
```

```python
#Se reliza el cambio de valores de la dimensión X4 y actualización dentro del dataframe dfIT2
dfIT2['X4']=dfIT2['X4'].apply(min_marr)
dfIT2['X4'].value_counts() 
```

```python
#Definir función para corrger los valores dentro de las columnas X6-X11
def rv_hist (value):
  if value<1:
    return -1
  else:
    return value
    
```

```python
#Aplicar la funcion de corrección dentro del dataframe
dfIT2['X6']=df['X6'].apply(rv_hist)
dfIT2['X7']=df['X7'].apply(rv_hist)
dfIT2['X8']=df['X8'].apply(rv_hist)
dfIT2['X9']=df['X9'].apply(rv_hist)
dfIT2['X10']=df['X10'].apply(rv_hist)
dfIT2['X11']=df['X11'].apply(rv_hist)
```
